```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../day1-data-cleaning/cleaned_btc_data.csv')

df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)
```

```python
daily = df.resample('D').mean()
```

```python
plt.figure(figsize=(14,6))
plt.plot(daily['close'])
plt.title('Bitcoin Price Trend', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
plt.show()
```

```python
daily['ma7'] = daily['close'].rolling(7).mean()
daily['ma30'] = daily['close'].rolling(30).mean()

plt.figure(figsize=(14,6))
plt.plot(daily['close'], label='Price')
plt.plot(daily['ma7'], label='MA 7')
plt.plot(daily['ma30'], label='MA 30')

plt.legend()
plt.title('Bitcoin Moving Average')
plt.grid()
plt.show()
```

```python
daily['returns'] = daily['close'].pct_change()

plt.figure(figsize=(14,6))
plt.plot(daily['returns'])
plt.title('Bitcoin Volatility (Returns)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
daily['close'].hist(bins=50)
plt.title('Distribution of Bitcoin Price')
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(daily['close'])
plt.yscale('log')
plt.title('Bitcoin Price (Log Scale)')
plt.grid()
plt.show()
```
